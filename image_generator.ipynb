```python
import torch
from torch import autocast
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
```

```python
authorization_token = "YOUR_AUTH_TOKEN"
```

```python
modelid = "CompVis/stable-diffusion-v1-4"
device = "cuda"
```

```python
pipe = StableDiffusionPipeline.from_pretrained(modelid, revision = "fp16", torch_dytpe = torch.float16, use_auth_token = authorization_token)
pipe.to(device)
```

```python
with autocast(device):
  textprompt = str(input("Enter your prompt: "))
  image = pipe(textprompt,guidance_scale = 8.5).images[0]
  imgplot = plt.imshow(image)
```
